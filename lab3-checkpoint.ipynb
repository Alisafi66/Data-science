```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn for countplot

# Create a sample DataFrame
data = {'incident_date': ['2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-04'],
        'other_column': [10, 15, 20, 25, 30]}

df = pd.DataFrame(data)

# Convert 'incident_date' to datetime if not already
df['incident_date'] = pd.to_datetime(df['incident_date'])

# Extract day of the week from the incident date
df['day_of_week'] = df['incident_date'].dt.day_name()

# Plotting the number of incidents reported on each day of the week
plt.figure(figsize=(8, 6))
sns.countplot(x='day_of_week', data=df)
plt.title('Number of Incidents Reported on Each Day of the Week')
plt.show()
```
